```python
import cv2
from IPython.display import Image
from matplotlib import pyplot as plt
from skimage.util import random_noise
```

# This is the original image that we'll work upon

```python
img=cv2.imread('original.jpg', -1)
Image('original.jpg')
```

## Gaussian noise

```python
gauss = random_noise(img, mode='gaussian', seed=None, clip=True)
plt.imsave('GAUSS.jpg', gauss)
Image('GAUSS.jpg')
```

## Localvar noise(type of gaussian noise)

```python
localvar = random_noise(img, mode='localvar', seed=None, clip=True)
plt.imsave('LOCALVAR.jpg', localvar)
Image('LOCALVAR.jpg')
```

## Poisson noise

```python
poison = random_noise(img, mode='poisson', seed=None, clip=True)
plt.imsave('POISSON.jpg', poison)
Image('POISSON.jpg')
```

## Salt noise

```python
salt = random_noise(img, mode='salt', seed=None, clip=True)
plt.imsave('SALT.jpg', salt)
Image('SALT.jpg')
```

## Pepper noise

```python
pepper = random_noise(img, mode='pepper', seed=None, clip=True)
plt.imsave('PEPPER.jpg', pepper)
Image('PEPPER.jpg')
```

## Speckle noise

```python
speckle = random_noise(img, mode='speckle', seed=None, clip=True)
plt.imsave('SPECKLE.jpg', speckle)
Image('SPECKLE.jpg')
```

## Salt & Pepper noise

```python
sp = random_noise(img, mode='s&p', seed=None, clip=True)
plt.imsave('S&P.jpg', sp)
Image('S&P.jpg')
```
